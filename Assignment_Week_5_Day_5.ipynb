## 2. Compare some Forecasting Techniques for Stationary and Non-stationary Time-Series.


###Forecasting Techniques for Stationary Time Series:

**Moving Average (MA) :-** This method calculates the average of a fixed-size window of past observations to forecast future values. It is useful for eliminating short-term fluctuations and identifying underlying patterns.

**Autoregressive (AR) :-** AR models use past observations to predict future values based on a linear combination of lagged variables. It assumes that the current value depends on previous values and a random error term.

**Autoregressive Moving Average (ARMA) :-** ARMA models combine the AR and MA models. They capture both the linear relationship with past observations (AR) and the residual errors (MA). ARMA models are useful for capturing both short-term and long-term dependencies in the data.

**Autoregressive Integrated Moving Average (ARIMA) :-** ARIMA models are suitable for forecasting stationary time series data with trends and seasonality. They combine differencing (integration) to make the data stationary with ARMA models to capture the autocorrelation structure.

###Forecasting Techniques for Non-Stationary Time Series:

**Trend Analysis :-** Non-stationary time series often exhibit trends, such as linear trends or exponential growth. Trend analysis techniques, such as linear regression or exponential smoothing, can be used to estimate and extrapolate the trend component.

**Seasonal Decomposition :-** Seasonal decomposition of time series (e.g., using the classical decomposition or X-12-ARIMA) separates the data into trend, seasonal, and residual components. These components can then be modeled and forecasted separately.

**Exponential Smoothing :-** Exponential smoothing methods, such as Simple Exponential Smoothing (SES) or Holt-Winters' methods, can be adapted for non-stationary time series by incorporating trend and seasonal components. These methods assign exponentially decreasing weights to past observations to generate forecasts.

**Vector Autoregression (VAR) :-** VAR models capture the interdependencies among multiple time series variables. They are useful when forecasting non-stationary time series that are influenced by each other. VAR models can capture both short-term and long-term relationships between variables.

**State Space Models :** State space models, including the popular Kalman filter and smoother algorithms, can be employed to model and forecast non-stationary time series. These models handle time-varying parameters and can capture complex dynamics in the data.

## 3. Why does a Time Series have to be Stationary?



We have seen that forecasting makes the assumption that the data is stationary. The model will not be able to faithfully reflect the underlying patterns in the data if it is non-stationary, and the findings will be erroneous.

The fact that non-stationarity might result in incorrect or misleading results is another justification for the necessity of time series being stationary before analysis. As an illustration, a non-stationary time series could give the impression that there is a high connection between two variables, but in reality, the correlation is only there because of a trend or a seasonal element in the data.

*  The Perceptron and Logistic Regression are both binary classification algorithms, but they differ in their underlying principles and the way they make predictions.
*   Logistic Regression is a statistical model that directly models the probability of the binary outcome using a logistic function. It is based on the concept of linear regression but uses a logistic (sigmoid) function to transform the output.

**Output Range**

* The Perceptron outputs binary class labels (+1 or -1) based on the sign of the linear combination of input features and weights.

*  The logistic function restricts the output between 0 and 1, representing the probability of belonging to the positive class.

**Decision Boundary :-**
* Decision boundary that separates the classes. It updates the weights of the perceptron iteratively to adjust the decision boundary until convergence.
