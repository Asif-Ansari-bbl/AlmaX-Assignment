The null hypothesis (H0) and alternative hypothesis (H1 or Ha) are two competing statements about a population parameter or the relationship between variables.

**Null Hypothesis (H0) :-** The null hypothesis represents the default or status quo assumption. It states that there is no significant difference or relationship between variables or no effect of a particular intervention. In other words, any observed difference or effect is due to random chance or sampling variability. The null hypothesis is typically denoted as "H0."

**Alternative Hypothesis (H1 or Ha) :-** The alternative hypothesis contradicts the null hypothesis. It states that there is a significant difference, relationship, or effect present in the population. The alternative hypothesis can take different forms depending on the specific research question or objective.

**One-tailed Tests**


*  A one-tailed test may be either left-tailed or right-tailed.

*  A left-tailed test is used when the alternative hypothesis states that the true value of the parameter specified in the null hypothesis is less than the null hypothesis claims.

*  A right-tailed test is used when the alternative hypothesis states that the true value of the parameter specified in the null hypothesis is greater than the null hypothesis claims.

## 1. What is Latent Dirichlet Allocation (LDA) and vector space of LDA ?

Latent Dirichlet Allocation (LDA) is a probabilistic model used for topic modeling, a technique used to discover hidden thematic structures in a collection of documents. LDA assumes that each document is a mixture of various topics, and each topic is characterized by a distribution of words.


The basic idea behind LDA is to represent documents as a mixture of topics and words as a distribution over topics. It assumes that documents are generated by a two-step process:

**Topic Assignment :-** Each document is composed of a mixture of topics. LDA assumes a Dirichlet prior over topic distributions in documents. For each document, a distribution over topics is sampled from the Dirichlet prior.

**Word Generation :-** Each word in a document is generated by first selecting a topic and then sampling a word from the topic's distribution over words. The topic for each word is sampled from the previously sampled distribution over topics.


The vector space representation of LDA involves transforming documents into a numerical representation that captures the distribution of topics within each document. This representation is typically a vector of topic proportions, where each dimension represents the weight of a particular topic in the document. These vectors can be used for various tasks such as document similarity, clustering, or classification.

In the vector space of LDA, each topic is represented by a distribution over words. These topic-word distributions provide insights into the most probable words associated with each topic. By examining the highest probability words for each topic, analysts can interpret and label the discovered topics based on the words that have the strongest association with them.

## 3. What are some advantages of Content-Based Recommendation paradigm over Collaborative-Based Recommendation?

Content-based filtering has many benefits compared to collaborative filtering, including:

*  No data from other users is required to start making recommendations. Unlike collaborative filtering, content-based filtering doesn’t need data from other users to create recommendations.


*  Recommendations are highly relevant to the user. Content-based recommenders can be highly tailored to the user’s interests, including recommendations for niche items, because the method relies on matching the characteristics or attributes of a database object with the user’s profile.


*  Recommendations are transparent to the user. Highly relevant recommendations project a sense of openness to the user, bolstering their trust level in offered recommendations. Comparatively, with collaborative filtering, instances are more likely to occur where users don’t understand why they see specific recommendations.

*  Collaborative filtering creates a potential cold start scenario when a new website or community has few new users and lacks user connections.  

## 4. What is a Model-Based Collaborative approach?

A Model-Based Collaborative approach, also known as Model-Based Collaborative Filtering, is a technique used in collaborative filtering-based recommendation systems. Collaborative filtering aims to generate recommendations by leveraging the similarities and patterns in user-item interactions or preferences. In Model-Based Collaborative filtering, a mathematical model is constructed based on the available user-item data to make predictions about user preferences for unseen items. Model-Based Collaborative approach leverages mathematical models to generate recommendations by capturing the underlying patterns and relationships in user-item data, offering advantages in scalability, accuracy, and flexibility compared to other collaborative filtering techniques.
