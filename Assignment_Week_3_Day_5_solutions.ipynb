## 1. a. What are the ways to avoid the problem of initialization sensitivity in the K means Algorithm?

## b. How to decide the optimal number of K in the K means Algorithm?



There are several way to avoid the problem of initialization sensitivity in K Mean Algorithm like 

*  **Repeat K Mean :-** It is basically repeat the algorithms again and again along with initilizing the centroids followed by picking up cluster which result in small intracluster distance and large intercluser distance.

*  **K Means++ :-** It is smart centroid initialization tecchnique. It is an improvement over random initialization. It selects the initial cluster centroids in a way that is more likely to lead to a good final clustering solution.

*  **Hierarchical Clustering Initialization :** Hierarchical clustering can be used as an initialization step for K-means. This approach can provide a better starting point by capturing the underlying structure of the data. 


We can decide the optimal number of K in K Mean Algorithm by following way like 

*  **Elbow Method :-** The Elbow Method is one of the most popular techniques for selecting K. It involves running K-means for a range of K values and calculating the sum of squared distances between data points and their cluster centroids. The idea is to plot the within-cluster sum of squares (WCSS) against the number of clusters K and identify the "elbow" point, where the rate of decrease in WCSS slows down significantly.

*  **Silhouette Coefficient :-** The Silhouette Coefficient is a measure of how well each data point fits into its assigned cluster compared to other clusters. It quantifies the compactness and separation of clusters. For different values of K, calculate the average Silhouette Coefficient across all data points. The K value that maximizes the average Silhouette Coefficient is considered the optimal number of clusters.

## 2. a. What is the Objective Function of k-Means?

## b. How would you perform k-Means on very large datasets?

The objective function of the k-means algorithm is to minimize the total within-cluster sum of squares (WCSS), also known as the total intra-cluster variance. The WCSS is a measure of how compact the clusters are. The objective of the k-means algorithm is to find the cluster assignments and the corresponding centroids that minimize the WCSS.

Performing k-means on very large datasets can be challenging due to memory and computational constraints. So we can apply these strategies to handle large datasets in k-means

* **Mini-Batch K-means :-** This approach reduces the memory requirements and computational complexity. Instead of using the entire dataset, you can use a random subset (mini-batch) of the data to update the cluster centroids in each iteration. This approach reduces the memory requirements and computational complexity. 

* **Sampling Techniques :-** If the dataset is too large to fit in memory, you can use sampling techniques to create a representative subset of the data. You can either perform random sampling or use more sophisticated techniques like stratified sampling to ensure that the samples cover the underlying distribution of the data.


* **Parallelization :-** Large-scale k-means can benefit from parallelization techniques. The data can be divided into subsets and processed in parallel on multiple processors or distributed computing frameworks like Apache Spark

* **Dimensionality Reduction :-** If the dataset has a high dimensionality, you can apply dimensionality reduction techniques like Principal Component Analysis (PCA) or t-SNE to reduce the number of features.

* **Incremental or Streaming K-means :-** For datasets that arrive in a streaming or online fashion, incremental or streaming k-means algorithms can be used. These algorithms update the centroids iteratively as new data points arrive, without requiring the entire dataset to be stored in memory.

## 3. a. Explain the different linkage methods used in the Hierarchical Clustering Algorithm.

## b. List down the pros and cons of complete and single linkages methods in the Hierarchical Clustering Algorithm.

There are three types of linkage used in Hierarchical Clustering like 

* **Single Linkage :-** In single linkage hierarchical clustering, the distance between two clusters is defined as the shortest distance between two points in each cluster. For example, the distance between clusters “A” and “B” to the left is equal to the length of the arrow between their two closest points.

 * <b> Pros of Single: </b> 
    *  This approach can separate non-elliptical shapes as long as the gap between two clusters is not small.
  * <b> Cons of Single: </b>

    *  MIN approach cannot separate clusters properly if there is noise between clusters.

    * It can cause premature merging of groups with close pairs, even if those groups are quite dissimilar overall. 


* **Complete Linkage :-** In complete linkage hierarchical clustering, the distance between two clusters is defined as the longest distance between two points in each cluster. For example, the distance between clusters “A” and “B” to the left is equal to the length of the arrow between their two furthest points.
    * <b>Pros of Complete :</b>

      * MAX approach does well in separating clusters if there is noise between clusters.
    * <b>Cons of Complete: </b>

      *  Max approach is biased towards globular clusters.
      *  Max approach tends to break large clusters.
* **Average Linkage :-**  In average linkage hierarchical clustering, the distance between two clusters is defined as the average distance between each point in one cluster to every point in the other cluster. For example, the distance between clusters “A” and “B” to the left is equal to the average length each arrow between connecting the points of one cluster to the other.
  * <b> Pros of Group Average:</b>

      * The group Average approach does well in separating clusters if there is noise between clusters.
  * <b> Cons of Group Average: </b>
    *  The group Average approach is biased towards globular clusters.

## 4. a. What is a dendrogram in Hierarchical Clustering Algorithm?

## b. Explain the different parts of dendrograms in the Hierarchical Clustering Algorithm.


A dendrogram is a graphical representation used in hierarchical clustering algorithms to visualize the clustering hierarchy and relationships among data points. A dendrogram is a tree-like structure that explains the relationship between all the data points in the system.


The dendrogram begins with all data points represented as individual leaves at the bottom. As the algorithm progresses, the most similar data points or clusters are progressively merged, and new branches are formed. The height of each branch in the dendrogram corresponds to the dissimilarity or distance between the clusters being merged.

There are different parts of dendrograms in the Hierarchical Clustering Algorithm like 


* **Leaves :-** The leaves of a dendrogram represent the individual data points in the dataset. Each data point is displayed as a separate leaf at the bottom of the dendrogram. 

* **Branches :-** The branches in a dendrogram represent the merging of clusters. At each level of the dendrogram, clusters are merged based on their similarity or distance.

* **Heights :-** The vertical axis of the dendrogram represents the dissimilarity or distance between clusters being merged.

* **Distance Measures :-**  The dendrogram visualizes the distances between data points or clusters based on the chosen metric. Common distance measures include Euclidean distance, Manhattan distance, correlation coefficient, etc.

* **Cutting Point :-** Dendrograms allow us to cut the tree at a specific height, creating different levels of clustering granularity. Cutting the dendrogram horizontally at a certain height results in separating the clusters into a desired number of groups.
