## 1. a. How are confidence tests and hypothesis tests similar? How are they different?

## b. What is the difference between type 1 error and type 2 error?


Confidence intervals and hypothesis tests are similar in that they are both inferential methods that rely on an approximated sampling distribution. Confidence intervals use data from a sample to estimate a population parameter. Hypothesis tests use data from a sample to test a specified hypothesis.


    Similarities:

    1. Both confidence tests and hypothesis tests are part of statistical inference, which involves drawing conclusions about a population based
    on sample data.

    2. Both tests involve the formulation of a null hypothesis.

    3. Both tests utilize the concept of sampling distributions.


    Differences:


    1. Confidence tests aim to estimate a range of plausible values for a population parameter and hypothesis tests aim to assess whether there is 
     sufficient evidence to reject the null hypothesis in favor of an alternative hypothesis.

    2. Confidence tests primarily focus on estimation, providing a range of values within which the true population parameter is likely to lie and 
     Hypothesis tests, on the other hand, are more concerned with decision-making that  determining whether the evidence supports
     rejecting or failing to reject the null hypothesis.

    3. Confidence tests are typically interpreted in terms of a range of values. For example, a 95% confidence interval for a population mean 
     would be interpreted as "We are 95% confident that the true population mean falls within this interval." Hypothesis tests, on the other hand, 
     provide a binary result: either reject the null hypothesis or fail to reject it.

    Type 1 Error:
    A Type 1 error, also known as a false positive, occurs when the null hypothesis (H0) is rejected when it is actually true.  
    Falsely predicted Positive

    Type 2 Error:
    A Type 2 error, also known as a false negative, occurs when the null hypothesis (H0) is not rejected when it is actually false. 
    Flasly predicted Negative

## 1. What is the difference between Classification and Regression when using SVM?

the main difference between classification and regression lies in the nature of the target variable and the goal of the modeling task.

**Classification with SVM :-**
Classification using SVM is a supervised learning task where the goal is to assign input data points to specific predefined categories or classes. In SVM classification, the target variable is categorical or discrete, representing different classes or labels. SVM seeks to find an optimal hyperplane that separates the data points of different classes with the maximum margin.

The primary objective of SVM classification is to construct a decision boundary that best separates the data into different classes.

**Regression with SVM :-**
Regression using SVM is also a supervised learning task, but it involves predicting a continuous or numerical target variable. In SVM regression, the goal is to find a function that best fits the data points and predicts the numerical output. SVM regression aims to find a hyperplane that approximates the relationship between the input features and the continuous target variable.

## 2. When SVM is not a good approach?

There are certain scenarios where SVM may not be the most suitable approach. 

**Large Datasets :-** SVMs can be computationally expensive, particularly when dealing with large datasets. As the number of training instances increases, the training time and memory requirements of SVM also increase significantly. In such cases, other algorithms like logistic regression or gradient boosting methods may be more efficient.

**High-Dimensional Data :-** SVMs can face challenges when working with high-dimensional data, where the number of features is much larger than the number of instances. This is known as the "curse of dimensionality." In high-dimensional spaces, SVMs may struggle to find an optimal hyperplane due to increased sparsity and overfitting concerns.

**Noisy Datasets :-** SVMs aim to find a clear decision boundary that maximizes the margin between classes. However, in the presence of significant noise or overlapping data points, SVMs might struggle to find an accurate boundary.

**Large Number of Classes :-** SVMs are inherently binary classifiers, designed to handle two-class classification problems. Extending SVMs to handle multi-class problems often involves techniques like one-vs-one or one-vs-all classification.

**Interpretability :-** SVMs generally provide good predictive performance, but they can lack interpretability, especially when using complex kernel functions or non-linear transformations

## 3. a.Explain learning rate in the context of neural network models. What happens if the learning rate is too high or too low?

## b. Explain the difference between a shallow network and a deep network.

 The learning rate is a hyperparameter that determines the step size at which the model's parameters are updated during the training process. It controls how quickly or slowly the model learns from the training data.

**When the learning rate is too high :-**

If the learning rate is set too high, it can lead to unstable and inefficient training. The model's parameter updates will be large, causing it to overshoot the optimal values. This can result in the loss function oscillating or even diverging, leading to poor convergence and inability to find the global minimum. In extreme cases, a high learning rate may prevent the model from converging at all.

**When the learning rate is too low :-**

If the learning rate is set too low, the model's parameter updates will be very small, making the training process slow and inefficient. The model may take a long time to converge, or it may get stuck in a suboptimal solution without reaching the global minimum.

A shallow network, also known as a shallow neural network, refers to a neural network architecture with a small number of hidden layers. Typically, shallow networks have only one or two hidden layers between the input and output layers. These networks are relatively simple in terms of their architecture and have fewer parameters compared to deep networks. Shallow networks are often used for simpler tasks or when the dataset is small and not highly complex.

A Deep network, also known as a deep neural network, is a neural network architecture that contains multiple hidden layers between the input and output layers. These networks have a significantly larger number of layers, typically ranging from three to dozens or even hundreds of layers. Deep networks are capable of learning hierarchical representations of data, allowing them to capture more complex patterns and relationships. Deep networks have become increasingly popular in various fields, such as computer vision, natural language processing, and speech recognition.

## 4. What are the different types of deep neural networks?

There are several types of deep neural networks, each designed to address specific tasks or architectural considerations like-

1. **Feedforward Neural Networks (FNN) :-**
Feedforward neural networks, also known as multilayer perceptrons (MLPs), are the basic type of deep neural networks. They consist of an input layer, one or more hidden layers, and an output layer. Information flows only in one direction, from the input layer through the hidden layers to the output layer. FNNs are used for various tasks, including classification, regression, and pattern recognition.

2. **Convolutional Neural Networks (CNN) :-**
Convolutional neural networks are specifically designed for processing structured grid-like data, such as images or audio. They employ convolutional layers that apply filters to local receptive fields, enabling them to capture spatial hierarchies and learn meaningful features from the input data. CNNs are widely used in computer vision tasks, such as image classification, object detection, and image segmentation.

3. **Recurrent Neural Networks (RNN) :-**
Recurrent neural networks are designed to handle sequential data, where the order and dependencies of the data elements matter, such as in natural language processing or time series analysis. RNNs have recurrent connections that allow information to flow not only from the input to the output but also across previous time steps within the network.

4. **Long Short-Term Memory Networks (LSTM):**
LSTM networks are a specialized type of recurrent neural network that addresses the vanishing gradient problem. They incorporate memory cells with gating mechanisms to selectively retain or discard information, making them effective in capturing long-term dependencies.

5. **Generative Adversarial Networks (GAN) :-**
Generative adversarial networks consist of two components: a generator and a discriminator. The generator learns to generate synthetic data samples that resemble real data, while the discriminator tries to distinguish between real and synthetic samples. Through adversarial training, GANs can generate high-quality, realistic samples, making them valuable for tasks such as image synthesis, text generation, and anomaly detection.
